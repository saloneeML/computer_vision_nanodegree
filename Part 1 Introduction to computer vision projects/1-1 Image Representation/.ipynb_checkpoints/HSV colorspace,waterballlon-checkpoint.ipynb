```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
%matplotlib inline
image = cv2.imread('waterballon.jpg')
```

```python
#Change color to RGB
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
#Display the image
plt.imshow(image)
```

```python
#RGB Channels
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]
f,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.set_title('red')
ax1.imshow(r,cmap='gray')
ax2.set_title('green')
ax2.imshow(g,cmap='gray')
ax3.set_title('blue')
ax3.imshow(b,cmap='gray')
              
```

```python
#Convert from RGB to Hsv
hsv = cv2.cvtColor(image,cv2.COLOR_RGB2HSV)
```

```python
#Convert from rgb to hue
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]
f,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))
ax1.set_title('Hue')
ax1.imshow(h,cmap='gray')
ax2.set_title('Saturation')
ax2.imshow(s,cmap='gray')
ax3.set_title('Value')
ax3.imshow(v,cmap='gray')
```

```python
#Define color selection criteria in HSV values
lower_hue = np.array([160,0,0])
upper_hue = np.array([180,255,255])
```

```python
#Define color selection criteria in RGB values
lower_pink = np.array([180,0,100])
upper_pink = np.array([255,255,230])
```

```python
#Define masked area in RGB Space
mask_rgb = cv2.inRange(image,lower_pink,upper_pink)
```

```python
#Mask the image
masked_image = np.copy(image)
masked_image[mask_rgb == 0] = [0,0,0]
```

```python
#Display mask
plt.imshow(masked_image)
```

```python
#Define masked area in HSV space
mask_hsv = cv2.inRange(hsv,lower_hue,upper_hue)
```

```python
#Mask the image
masked_image = np.copy(image)
masked_image[mask_hsv == 0] = [0,0,0]
```

```python
#Visualize the mask
plt.imshow(masked_image)
```
